#### 4.Relation

```python
df["Relation"]=df["Marital_Status"].replace({"Married":"Relation", "Together":"Relation", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone"})
```

```python
# Extract marital status data from DataFrame
marital_status_data = df['Relation'].value_counts().to_dict()

# Declare data lists
data = list(marital_status_data.values())
keys = list(marital_status_data.keys())

# Declare exploding pie
explode = [0, 0.1]

# Define Seaborn color palette to use
palette_color = sns.color_palette('colorblind')

# Plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')

# Displaying chart
plt.title(" Marital Status Distribution")
plt.show()
```
